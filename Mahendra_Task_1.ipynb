```python
# Run this cell in Google Colab

import pandas as pd
import matplotlib.pyplot as plt

# Sample data (you can replace this with actual data from World Bank or other source)
# Gender distribution (categorical - bar chart)
gender_data = {
    'Gender': ['Male', 'Female'],
    'Population': [500000, 520000]
}

# Age distribution (continuous - histogram)
age_data = {
    'Age': [5, 12, 15, 18, 22, 25, 30, 35, 40, 45, 50, 60, 65, 70, 75, 80, 85, 90, 95, 100]
}

# Convert to DataFrames
df_gender = pd.DataFrame(gender_data)
df_age = pd.DataFrame(age_data)

# --- Bar Chart: Gender Distribution ---
plt.figure(figsize=(6,4))
plt.bar(df_gender['Gender'], df_gender['Population'], color=['skyblue', 'lightpink'])
plt.title('Population by Gender')
plt.xlabel('Gender')
plt.ylabel('Population')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# --- Histogram: Age Distribution ---
plt.figure(figsize=(8,5))
plt.hist(df_age['Age'], bins=10, color='lightgreen', edgecolor='black')
plt.title('Age Distribution in Population')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
